## Adding title

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
iris = sns.load_dataset('iris')

sns.lineplot(x = "sepal_length", y="sepal_width", data=iris)
plt.title("Iris plot")
plt.show
```

## Adding limits

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
iris = sns.load_dataset('iris')

sns.lineplot(x = "sepal_length", y="sepal_width", data=iris)
plt.title("Iris plot")
plt.xlim(1)
plt.ylim(1)
```

## Set Styles
- Darkgrid
- Whitegrid
- Dark
- White
- Ticks


```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style(style = None, rc = None)
iris = sns.load_dataset("iris")
sns.lineplot(x = "sepal_length", y = "sepal_width", data=iris)
sns.set_style("darkgrid")
plt.title("Phoolon ka set")
plt.show()
```

## Size of Figure

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

iris = sns.load_dataset("iris")

sns.lineplot(x = "sepal_width", y = "sepal_length", data = iris)
plt.title("Phoolon ka title")
plt.show()
```
